Data Splitting


```python
X=data.drop('diagnosis',axis=1)
y=data['diagnosis']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.33,random_state=42)
```

Standardizing the data


```python
from sklearn.preprocessing import StandardScaler
std=StandardScaler()
X= std.fit(X).transform(X)
X_trainstd=std.fit_transform(X_train)
X_teststd=std.transform(X_test)
```

Creating the Neural Network


```python
import tensorflow as tf
tf.random.set_seed(42)
from tensorflow import keras
```

```python
model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20,activation='relu'),
                          keras.layers.Dense(2,activation= 'sigmoid')
])
```

Compiling the Neural Network

```python
model.compile(optimizer="adam",
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
     
```

```python
history=model.fit(X_trainstd,y_train,validation_split=0.1,epochs=10)
```

```python
loss,accuracy=model.evaluate(X_teststd,y_test)
print(accuracy)
```

Making a predictive system


```python
print(X_teststd)
```

```python
#building a predictive system
input=(13.49,22.3,86.91,561,0.08752,0.07698,0.04751,0.03384,0.1809,0.05718,0.2338,1.353,1.735,20.2,0.004455,0.01382,0.02095,0.01184,0.01641,0.001956,15.15,31.82,99,698.8,0.1162,0.1711,0.2282,0.1282,0.2871,0.06917)
#Changing the inout data to a numpy array
input_array =np.asarray(input)
#Reshaping the input array as we are predicting for one data point
input_reshaped = input_array.reshape(1,-1)
#Standardising the reshaped input data
input_Std = std.transform(input_reshaped)
#Prediction
prediction = model.predict(input_Std)
print(prediction)
prediction_label= [np.argmax(prediction)]
print(prediction_label)
if (prediction_label[0]==0 ):
    print ('The Tumor is Benign')
else:
    print('The Tumor is Malignant' )
```
